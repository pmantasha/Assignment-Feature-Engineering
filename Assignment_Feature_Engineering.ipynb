#Theory:-

1. What is a parameter?
   - A parameter is an internal variable of the machine learning model that is learned from the training data, such as weights in a linear regression or neural network.

2. What is correlation? What does negative correlation mean?
   - Correlation measures the strength and direction of a linear relationship between two variables.

   - Negative correlation means that as one variable increases, the other decreases. For example, the relationship between exercise frequency and body weight.

3. Define Machine Learning. What are the main components in Machine Learning?  
   - Machine Learning is a method where machines learn from data without being explicitly programmed.

   - Main components:

    - Dataset

    - Features

    - Model

    - Loss function

    - Optimizer

    - Evaluation metrics

4. How does loss value help in determining whether the model is good or not?
   - The loss value indicates how well the model's predictions match the actual values. A lower loss means a better fit; a high loss suggests poor performance.

5. What are continuous and categorical variables?
   - Continuous variables: Can take any numerical value (e.g., height, weight).

   - Categorical variables: Represent categories or labels (e.g., gender, city).

6. How do we handle categorical variables in Machine Learning? What are the common techniques?
  - We convert categorical variables into numerical form using:

    - Label Encoding

    - One-Hot Encoding

    - Ordinal Encoding

7. What do you mean by training and testing a dataset?
   - Training dataset: Used to train the model.

   - Testing dataset: Used to evaluate model performance on unseen data.

8. What is sklearn.preprocessing?
  - sklearn.preprocessing is a module in Scikit-learn that provides functions to scale, normalize, and encode features.

9. What is a Test set?
  - The test set is a separate portion of the dataset used to evaluate the model’s performance after training.

11. Why do we have to perform EDA before fitting a model to the data?
   - EDA helps in understanding the structure, patterns, outliers, and relationships in the data to prepare it for accurate model training.

15. What is causation? Explain difference between correlation and causation with an example.
  - Causation: Implies that one variable directly affects the other.

  - Correlation: Implies a relationship but not cause and effect.

   - Example:
    Ice cream sales and drowning deaths are correlated, but the cause is summer heat, not ice cream.

16. What is an Optimizer? What are different types of optimizers? Explain each with an example.
  - Optimizers adjust model parameters to minimize the loss function.

  - Types:

     - Gradient Descent: Standard optimizer.

     - Stochastic Gradient Descent (SGD): Uses random samples.

     - Adam Optimizer: Combines momentum and adaptive learning.

   - Example: optimizer = tf.keras.optimizers.Adam()

17. What is sklearn.linear_model?
  - sklearn.linear_model is a Scikit-learn module that provides algorithms like Linear Regression and Logistic Regression.

18. What does model.fit() do? What arguments must be given?
  - model.fit() trains the model.

  - Arguments:

     - Training data (X_train)

     - Target labels (y_train)

19. What does model.predict() do? What arguments must be given?
  - model.predict() makes predictions on new data.

  - Argument:

     - Input data for prediction.

20. What are continuous and categorical variables?
  - Continuous: Numeric values that can be measured (e.g., temperature).

  - Categorical: Distinct categories (e.g., color: red, blue, green).

21. What is feature scaling? How does it help in Machine Learning?
  - Feature scaling standardizes or normalizes data so that all features contribute equally to the model.

  - It helps:

     - Improve model convergence.

     - Improve accuracy in distance-based algorithms.
